```python
fields = list(allMatchData.columns[2:-2])
finalfields = [ x.replace("team_0","") for x in fields if "team_1" not in x ]
finalfields =finalfields[:-1]
len(finalfields)
```

```python
mergeaddfieldnames = ['matchid','team_0_won_before','team_1_won_before'];
for curfield in finalfields:
    mergeaddfieldnames.append('team_0_lastwinmatch_'+curfield)
    mergeaddfieldnames.append('team_1__lastwinmatch_'+curfield)    
```

```python
len(mergeaddfieldnames)
```

```python
sourceDf = pd.merge(eventsMlData,mlMatchData[mergeaddfieldnames],on='matchid', how='left')
```

```python
sourceDf.dtypes
```

```python
#remove fields
removefields = ['id','matchid','matchid','timestamp','team','winner']
topredict='winner'
mlInput = sourceDf.drop(removefields, axis=1)
```

```python
mlInput.head()
```

```python
from sklearn.feature_extraction import DictVectorizer
v = DictVectorizer(sparse=False)
X = v.fit_transform(mlInput.to_dict(orient='records'))
from sklearn.preprocessing import Normalizer
transformer = Normalizer().fit(X)
Y=transformer.transform(X)
```

```python
from sklearn import preprocessing
classesToTransfrom= list(sourceDf[topredict])
le = preprocessing.LabelEncoder()
classes = le.fit_transform(classesToTransfrom)
```

```python
X.shape
```

```python
classes.shape
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(max_iter=400,random_state=0, solver='lbfgs',multi_class='ovr').fit(X, classes)
```

```python
probs=clf.predict_proba(X)
```

```python
classLabels =le.inverse_transform(clf.classes_)
```

```python
#add prediction back to source dataset
dfpred = pd.DataFrame(probs, columns=classLabels)
```

```python
for label in classLabels:
    dfpred[label] = dfpred[label].apply(lambda x: round(x, 4)*100)
    sourceDf['win_perc_'+label]=dfpred[label]
```
